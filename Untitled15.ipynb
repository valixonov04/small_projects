```python
import pandas as pd
import numpy as np
import sklearn
```

```python
df = pd.read_csv("https://raw.githubusercontent.com/anvarnarz/praktikum_datasets/refs/heads/main/diabetes.csv")
df
```

##Ma'lumotlar ustida ishlash

```python
df.info()
```

```python
#corrlatsiya
corr_matrix = df.corr().abs()
corr_matrix.style.background_gradient(cmap="GnBu")
```

```python
df.Outcome.value_counts()
```

```python
index_out = df[df['Outcome'] == 0].sample(250).index
df.drop(index_out,inplace=True, axis=0)
df
```

```python
df.info()
```

```python
df.corrwith(df.Outcome)
```

```python
X = df.drop("Outcome",axis=1).values
y = df.Outcome
```

```python
from sklearn.preprocessing import StandardScaler
ss = StandardScaler()
X = ss.fit_transform(X)
```

```python
from inspect import trace
from sklearn.model_selection import train_test_split
X_train ,X_test, y_train ,y_test = train_test_split(X,y, train_size=0.9 , random_state=42)
```

```python
y_train.shape
```

##ML

```python
from sklearn.neighbors import KNeighborsClassifier
knn = KNeighborsClassifier(n_neighbors=3) #k - tahminiy
knn.fit(X_train, y_train)
```

```python
predict_knn = knn.predict(X_test)
```

```python
from sklearn.metrics import jaccard_score
jaccard_score(y_test,predict_knn)
```

```python
from sklearn.metrics import confusion_matrix
import seaborn as sns
import matplotlib.pyplot as plt
sns.heatmap(confusion_matrix(y_test,predict_knn),annot = True)
plt.show()
```

```python
confusion_matrix(y_test, predict_knn)
```

```python
from sklearn.metrics import classification_report
cr = classification_report(y_test, predict_knn)
print(cr)
```

```python
from sklearn.model_selection import cross_val_predict
predict = cross_val_predict(estimator= knn,X=X,y=y,cv=5)
from sklearn.metrics import classification_report
cr = classification_report(y, predict)
print(cr)
```

```python
from sklearn.metrics import f1_score
f1 = []
for k in range(1,25):
  knn = KNeighborsClassifier(n_neighbors=k)
  knn.fit(X_train,y_train)
  y_predict = knn.predict(X_test)
  f1.append(f1_score(y_test,y_predict))

plt.figure(figsize=(10,6))
plt.plot(range(1,25),f1)
plt.xticks(range(1,25))
plt.grid()
plt.show()
```
