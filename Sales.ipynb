```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("https://raw.githubusercontent.com/anvarnarz/praktikum_datasets/refs/heads/main/reklama_sales.csv")
```

```python
df
```

```python
df.corrwith(df.Sales)
```

```python
plt.figure(figsize=(12,6))
plt.subplot(1,3,1)
plt.scatter(data=df , y="Newspaper",x="Sales")
plt.title("Newspaper")
plt.subplot(1,3,2)
plt.scatter(data=df , y="TV",x="Sales")
plt.title("TV")
plt.subplot(1,3,3)
plt.scatter(data=df , y="Radio",x="Sales")
plt.title("Radio")
plt.show()
```

```python
#chizmadan ko'rinib turibdiki TV da berilgan reklamalar ancha sotuvlarga tasir qilgan
from sklearn.model_selection import train_test_split
train_set,test_set = train_test_split(df, random_state=42,train_size=0.9)
```

```python
train_set
```

```python
test_set
```

```python
X = train_set.drop("Sales", axis=1)
y = train_set.Sales
```

```python
X_num = np.asanyarray(X)
y_num = np.asanyarray(y)
```

```python
X_num
```

```python
from sklearn.linear_model import LinearRegression
LR_model = LinearRegression()
LR_model.fit(X_num,y_num)
```

```python
X_test_num = np.asanyarray(test_set.drop("Sales",axis=1))
y_test_num = np.asanyarray(test_set.Sales)
```

```python
predict_model = LR_model.predict(X_test_num)
predict_model
```

```python
y_test_num
```

```python
from sklearn.metrics import mean_squared_error
mse = mean_squared_error(y_test_num,predict_model)
print("MSE: ",mse )
```

```python
mse = mean_squared_error(y_test_num,predict_model)
print("RMSE: ",np.sqrt(mse))
```
